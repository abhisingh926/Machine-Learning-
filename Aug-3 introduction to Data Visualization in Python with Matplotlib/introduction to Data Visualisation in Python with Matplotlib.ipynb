```python
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
```

```python
#Basic Graph
x=[1,2,3]
y=[3,5,6]
plt.figure(figsize=(5,3),dpi=200)

plt.plot(x,y, label='2X', color='red', marker='*')
plt.title('first grap', fontdict={'fontname':'Comic Sans MS'})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

x2 = np.arange(0,4.5,0.5)
plt.plot(x2, x2**2)

plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])

plt.savefig('1st-graph.png', dpi = 300)

plt.legend()
plt.show() 
```

```python
labels = ['A','B','C']
values = [1,4,8]


bars = plt.bar(labels,values)

patterns = ['/','','*']
for bar in bars:
    bar.set_hatch(patterns.pop())

# bars[0].set_hatch('/')
plt.show()
```
