```python
import pandas as pd
df =pd.read_csv('import_csv/85_Imp_Jun-21.csv', encoding = "ISO-8859-1")
code = [85414011,85414012]
df = df[df['HS_CODE'].isin(code)]
df.head()
```

```python
df
```

```python
wp_range = list(range(300,600,5))
WP = list(range(300,600,5))
WP = list(map(str, WP))
wp1 = pd.concat([df['PRODUCT_DESCRIPTION'].str.extract('(?P<WP>{})'.format("|".join(WP)), expand=True).fillna('0')], axis=1)
df['wp']= wp1
df['IMP_DATE'] = pd.to_datetime(df['IMP_DATE'], format='%d-%b-%y')
df
```

```python
df['SUPPLIER_NAME'] = df['SUPPLIER_NAME'].str.replace('[^\w\s]','')
df['IMPORTER_NAME'] = df['IMPORTER_NAME'].str.replace('[^\w\s]','')
d1 = pd.read_csv('shipper/shipper_jul.csv')
d1['name'] = d1['name'].str.replace('[^\w\s]','')
d11= pd.Series(d1.id.values,index=d1.name).to_dict()
df['supplier_id'] = df['SUPPLIER_NAME'].map(d11)
```
